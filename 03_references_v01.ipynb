#### all()

Mengembalikan True jika semua item dalam objek iterable benar.


```python
# Mengecek apakah semua items di dalam list adalah True

mylist = [True, True, True]
x = all(mylist)
print(x)
```

Fungsi **all()** mengembalikan **True** jika semua item dalam iterable benar, jika tidak maka akan mengembalikan **False**.

Jika objek iterable kosong, fungsi **all()** juga mengembalikan **True**.


Syntax nya adalah **all(iterable)**


image.png

```python
# Mengecek apakah semua items di dalam list adalah True

mylist = [0, 1, 1]
x = all(mylist)
print(x)

# Mengembalikan False karena 0 adalah sama dengan False
```

```python
# Mengecek apakah semua items di dalam tuple adalah True

mytuple = (0, True, False)
x = all(mytuple)
print(x)

# Mengembalikan False karena baik item pertama dan ketiga adalah False
```

```python
# Mengecek apakah semua items di dalam set adalah True

myset = {0, 1, 0}
x = all(myset)
print(x)

# Mengembalikan False karena nilai item pertama dan ketiga adalah False
```

```python
# Mengecek apakah semua items di dalam dictionary adalah True

mydict = {0 : "Apple", 1 : "Orange"}
x = all(mydict)
print(x)

# Mengembalikan False karena KEY pertama adalah False.

# Untuk tipe data dictionary, fungsi all() akan mengecek KEY, bukan VALUE (nilai) nya
```

**Catatan:** Saat digunakan pada dictionary, fungsi **all()** memeriksa apakah semua KEY benar, bukan VALUE nya.

#### compile()

Mengembalikan sumber yang ditentukan sebagai objek, siap dieksekusi.


```python
# Kompilasi teks sebagai kode, dan jalankan

x = compile('print(55)', 'test', 'eval')
exec(x)
```

Fungsi **compile()** mengembalikan sumber yang ditentukan sebagai objek kode, siap untuk dieksekusi.


Syntax nya adalah **compile(source, filename, mode, flag, dont_inherit, optimize)**


Penjelasan parameter:

**sumber** Diperlukan. Sumber yang akan dikompilasi, bisa berupa String, objek Bytes, atau objek AST

**filename** Diperlukan. Nama file yang sumbernya berasal. Jika sumbernya bukan dari file, kita dapat menulis apa pun yang kita suka

**mode** Diperlukan. Nilai hukum:
* eval - jika sumbernya adalah ekspresi tunggal
* exec - jika sumbernya adalah blok pernyataan
* single - jika sumbernya adalah pernyataan interaktif tunggal

**flag** Opsional. Bagaimana mengkompilasi sumber. Standar 0

**dont-inherit** Opsional. Bagaimana mengkompilasi sumber. Default False

**optimize** Opsional. Menentukan tingkat optimalisasi kompiler. Default -1


```python
# Kompilasi lebih dari 1 statement dan eksekusi

x = compile('print(55)\nprint(88)', 'test', 'exec')
exec(x)
```

#### filter()

Gunakan fungsi filter untuk mengecualikan item dalam objek yang dapat diubah


```python
# Filter sebuah array, dan kembalikan sebuah new array array dengan nilai hanya >= 18

ages = [5, 12, 17, 18, 24, 32]

def myFunc(x):
  if x < 18:
    return False
  else:
    return True

adults = filter(myFunc, ages)

for x in adults:
  print(x)
```

Fungsi **filter()** mengembalikan iterator jika item difilter melalui fungsi untuk menguji apakah item diterima atau tidak.


Syntax nya adalah **filter(function, iterable)**


Parameter **function**:

Fungsi yang akan dijalankan untuk setiap item dalam iterable


Parameter **iterable**:

Iterable untuk difilter

#### issubclass()

Mengembalikan **True** jika class tertentu adalah subclass dari objek tertentu.


```python
# Periksa jika class myObj adalah subclass dari myAge

class myAge:
  age = 36

class myObj(myAge):
  name = "John"
  age = myAge

x = issubclass(myObj, myAge)

print(x)
```

Fungsi **issubclass()** mengembalikan **True** jika objek yang ditentukan adalah subclass dari objek yang ditentukan, jika tidak, **False**.

Syntax nya adalah **issubclass(object, subclass)**


Parameter **object**:

Dibutuhkan. Merupakan sebuah objek.


Parameter **subclass**:

Sebuah objek class, atau sebuah tuple dari objek class.

#### map()

Mengembalikan iterator yang ditentukan dengan fungsi yang ditentukan dan diterapkan ke setiap item


```python
# Hitung (kalkulasikan) panjang dari setiap kata pada tubpe

def myfunc(a):
  return len(a)

x = map(myfunc, ('apple', 'banana', 'cherry'))

print(x)

# Konversikan (ubah) map menjadi list, agar mudah (bisa) dibaca
print(list(x))
```

Fungsi **map()** mengeksekusi fungsi yang ditentukan untuk setiap item dalam iterable. Item dikirim ke fungsi sebagai parameter.


Syntax nya adalah **map(function, iterables)**


Parameter **function**:

Diperlukan. Fungsi untuk mengeksekusi untuk setiap item.


Parameter **iterable**:

Diperlukan. Urutan, koleksi, atau objek iterator. Kita dapat mengirim iterable sebanyak yang kita suka, pastikan fungsinya memiliki satu parameter untuk setiap iterable.


```python
# Buatlah daftar (list) buah baru, dengan mengirimkan 2 objek iterable kedalam fungsi

def myfunc(a, b):
  return a + b

x = map(myfunc, ('apple', 'banana', 'cherry'), ('orange', 'lemon', 'pineapple'))

print(x)

#convert the map into a list, for readability:
print(list(x))
```

#### range()

Mengembalikan urutan angka, mulai dari 0 dan bertambah 1 (secara *default*).


```python
# Buat urutan angka dari 0 hingga 5, dan cetak setiap item dalam urutan tersebut

x = range(6)
for n in x:
  print(n)
```

Fungsi **range()** mengembalikan urutan angka, mulai dari 0 secara default, dan bertambah 1 (secara default), dan berhenti sebelum angka yang ditentukan.


Syntax nya adalah **range(start, stop, step)**


Parameter **start** adalah Opsional. Angka bilangan bulat yang menentukan di mana posisi untuk memulai. Standarnya (*default*) adalah 0.


Parameter **stop** adalah Diperlukan. Angka bilangan bulat yang menentukan posisi berhenti (tidak disertakan).


Parameter **step** adalah Opsional. Angka bilangan bulat yang menentukan penambahan. Standarnya (*default*) adalah 1.


```python
# Buat sebuah urutan angka dari 3 sampai 5 dan cetak masing-masing item dalam urutan tersebut

x = range(3, 6)
for n in x:
  print(n)
```

```python
# Buat sebuah urutan angka dari 3 sampai 19, tapi penambahan nya 2

x = range(3, 20, 2)
for n in x:
  print(n)
```

#### slice()

Mengembalikan objek irisan.


```python
# Buat tuple dan objek irisan. Gunakan objek slice untuk mendapatkan hanya dua item pertama dari tuple

a = ("a", "b", "c", "d", "e", "f", "g", "h")

x = slice(2)

print(a[x])
```

Fungsi **slice()** mengembalikan objek slice.

Objek irisan digunakan untuk menentukan cara mengiris urutan. Kita dapat menentukan di mana memulai pemotongan, dan di mana harus mengakhiri. Kita juga dapat menentukan langkahnya, yang memungkinkan kita untuk misalnya: iris hanya setiap item lainnya.


Syntax nya adalah **slice(start, end, step)**


Parameter **start**: Opsional dan merupakan angka bilangan bulat yang menentukan di posisi mana untuk memulai pemotongan. Standarnya (*default*) adalah 0.

Parameter **end**: merupakan angka integer yang menentukan di posisi mana untuk mengakhiri pemotongan

Parameter **step**: Opsional dan merupakan bilangan bulat yang menentukan langkah pemotongan. Standarnya (*default*) adalah 1.


```python
# Buat tuple dan objek irisan
# Mulai objek irisan di posisi 3, dan iris ke posisi 5, dan kembalikan hasilnya

a = ("a", "b", "c", "d", "e", "f", "g", "h")
x = slice(3, 5)
print(a[x])
```

```python
# Buat tuple dan objek irisan
# Gunakan parameter langkah untuk mengembalikan setiap item ketiga

a = ("a", "b", "c", "d", "e", "f", "g", "h")
x = slice(0, 8, 3)
print(a[x])
```

#### sorted()

Mengembalikan daftar yang diurutkan.

```python
# Sort descending

a = ("h", "b", "a", "c", "f", "d", "e", "g")
x = sorted(a, reverse=True)
print(x)
```

#### super()

Mengembalikan objek yang mewakili kelas induk (*parent class*).


```python
# Buat class yang akan mewarisi semua metode dan properti dari class lain

class Parent:
  def __init__(self, txt):
    self.message = txt

  def printmessage(self):
    print(self.message)

class Child(Parent):
  def __init__(self, txt):
    super().__init__(txt)

x = Child("Hello, and welcome!")

x.printmessage()
```

Fungsi **super()** digunakan untuk memberikan akses ke metode dan properti dari kelas induk (*parent class*) atau saudara kandung (*sibling class*).

Fungsi **super()** mengembalikan objek yang mewakili kelas induk (*parent class*).


Syntax nya adalah **super()**


Tidak ada parameter yang diperlukan.

#### zip()

Mengembalikan sebuah iterator, dari dua atau lebih iterator.


```python
# Gabungkan dua tuple bersama

a = ("John", "Charles", "Mike")
b = ("Jenny", "Christy", "Monica")

x = zip(a, b)

# gunakan fungsi tuple() untuk menampilkan versi hasil yang dapat dibaca

print(tuple(x))
```

Fungsi **zip()** mengembalikan objek zip, yang merupakan iterator dari tupel di mana item pertama di setiap iterator yang diteruskan dipasangkan bersama, dan kemudian item kedua di setiap iterator yang diteruskan dipasangkan bersama, dll.

Jika iterator yang diteruskan memiliki panjang yang berbeda, iterator dengan item paling sedikit akan menentukan panjang iterator baru.


Syntax nya adalah **zip(iterator1, iterator2, iterator3 ...)**


Parameter **iterator1, iterator2, iterator3 ...** adalah objek iterator yang akan digabungkan bersama


```python
# Jika satu tuple berisi lebih banyak item, item ini akan diabaikan

a = ("John", "Charles", "Mike")
b = ("Jenny", "Christy", "Monica", "Vicky")

x = zip(a, b)

# Gunakan fungsi tuple() untuk menampilkan versi hasil yang dapat dibaca

print(tuple(x))
```
